```python
def display(board):
    for row in board:
        print(row)

def isSafe(board, x, y, n):
    """
    returns if placing Queen is viable or not

    args:
    x (int): row to be placed in
    y (int): col to be placed in
    n (int): board size

    return:
    bool: returns is it Safe or not?
    """
    # Column Test
    for row in range(x):
        if (board[row][y]=='Q'):
            return False

    # Top left diagonal test
    row = x - 1
    col = y - 1
    while row >=0 and col >=0:
        if (board[row][col] == 'Q'):
            return False
        row -= 1
        col -= 1

    # Top Right Diagonal Test
    row = x - 1
    col = y + 1
    while row >= 0 and col < n:
        if (board[row][col] == 'Q'):
            return False
        row -= 1
        col += 1
    return True

def nQueensSolver(board, x, n):
    # if nQueens are placed returns True
    if (x>=n):
        return True
    # interate through columns for each row
    for col in range(n):
        # if the pos is safe then place Queen
        if(isSafe(board, x, col, n)):
            board[x][col] = 'Q'
        # if next Queen can be placed return true
            if(nQueensSolver(board, x+1, n)):
                return True
        # else backtrack
        board[x][col] = ' '
    return False
```

```python
import tkinter as tk
from tkinter import simpledialog

def display_solution(board):
    window = tk.Tk()
    window.title("n-Queens Solution")

    for i in range(len(board)):
        for j in range(len(board)):
            cell = tk.Label(window, text=board[i][j], borderwidth=1, width=4, height=2, relief="solid")
            cell.grid(row=i, column=j)

    window.mainloop()

def display(board):
    for row in board:
        print(' '.join(row))

def isSafe(board, x, y, n):
    for row in range(x):
        if board[row][y] == 'Q':
            return False

    row, col = x - 1, y - 1
    while row >= 0 and col >= 0:
        if board[row][col] == 'Q':
            return False
        row -= 1
        col -= 1

    row, col = x - 1, y + 1
    while row >= 0 and col < n:
        if board[row][col] == 'Q':
            return False
        row -= 1
        col += 1
    return True

def nQueensSolver(board, x, n):
    if x >= n:
        return True
    for col in range(n):
        if isSafe(board, x, col, n):
            board[x][col] = 'Q'
            if nQueensSolver(board, x + 1, n):
                return True
            board[x][col] = ' '
    return False

def main():
    root = tk.Tk()
    root.withdraw()  # Hide the root window

    n = simpledialog.askinteger("Input", "Enter n:", minvalue=1)
    board = [[' '] * n for _ in range(n)]

    if nQueensSolver(board, 0, n):
        display(board)
        display_solution(board)
    else:
        print("No Solution")

    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import simpledialog

def display_solution(board):
    window = tk.Tk()
    window.title("n-Queens Solution")

    for i in range(len(board)):
        for j in range(len(board)):
            cell = tk.Label(window, text=board[i][j], borderwidth=1, width=4, height=2, relief="solid")
            cell.grid(row=i, column=j)

    window.mainloop()

def display(board):
    for row in board:
        print(' '.join(row))

def isSafe(board, x, y, n):
    for row in range(x):
        if board[row][y] == 'Q':
            return False

    row, col = x - 1, y - 1
    while row >= 0 and col >= 0:
        if board[row][col] == 'Q':
            return False
        row -= 1
        col -= 1

    row, col = x - 1, y + 1
    while row >= 0 and col < n:
        if board[row][col] == 'Q':
            return False
        row -= 1
        col += 1
    return True

def nQueensSolver(board, x, n, level=0):
    if x >= n:
        return True
    for col in range(n):
        print(f"{' ' * (level * 2)}Trying to place Queen at ({x}, {col})")
        if isSafe(board, x, col, n):
            board[x][col] = 'Q'
            print(f"{' ' * (level * 2)}Placed Queen at ({x}, {col})")
            if nQueensSolver(board, x + 1, n, level + 1):
                return True
            board[x][col] = ' '
            print(f"{' ' * (level * 2)}Backtracking from ({x}, {col})")
    return False

def main():
    root = tk.Tk()
    root.withdraw()

    n = simpledialog.askinteger("Input", "Enter n:", minvalue=1)
    board = [[' '] * n for _ in range(n)]

    if nQueensSolver(board, 0, n):
        display(board)
        display_solution(board)
    else:
        print("No Solution")

    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
import time

class TowerOfHanoi:
    def __init__(self, root, num_discs):
        self.root = root
        self.num_discs = num_discs
        self.canvas = tk.Canvas(root, width=600, height=400)
        self.canvas.pack()
        self.rods = [[i for i in range(num_discs, 0, -1)], [], []]
        self.discs = []
        self.create_rods()
        self.create_discs()
        self.animate_hanoi(num_discs, 0, 2, 1)
    
    def create_rods(self):
        rod_width = 10
        rod_height = 200
        base_height = 20
        self.canvas.create_rectangle(100 - rod_width//2, 400 - rod_height, 100 + rod_width//2, 400, fill="black")
        self.canvas.create_rectangle(300 - rod_width//2, 400 - rod_height, 300 + rod_width//2, 400, fill="black")
        self.canvas.create_rectangle(500 - rod_width//2, 400 - rod_height, 500 + rod_width//2, 400, fill="black")
        self.canvas.create_rectangle(50, 400, 550, 400 + base_height, fill="black")
    
    def create_discs(self):
        disc_height = 20
        max_disc_width = 200
        min_disc_width = 50
        for i in range(self.num_discs):
            width = min_disc_width + (max_disc_width - min_disc_width) * (self.num_discs - i) // self.num_discs
            disc = self.canvas.create_rectangle(100 - width//2, 400 - 20 * (i + 1), 100 + width//2, 400 - 20 * i, fill="blue")
            self.discs.append(disc)
    
    def move_disc(self, from_rod, to_rod):
        disc = self.rods[from_rod].pop()
        self.rods[to_rod].append(disc)
        self.update_canvas()
    
    def update_canvas(self):
        rod_positions = [100, 300, 500]
        for rod_index, rod in enumerate(self.rods):
            for disc_index, disc in enumerate(rod):
                x = rod_positions[rod_index]
                y = 400 - 20 * (disc_index + 1)
                disc_width = self.canvas.coords(self.discs[disc - 1])[2] - self.canvas.coords(self.discs[disc - 1])[0]
                self.canvas.coords(self.discs[disc - 1], x - disc_width // 2, y - 20, x + disc_width // 2, y)
        self.root.update()
        time.sleep(0.5)
    
    def animate_hanoi(self, n, from_rod, to_rod, aux_rod):
        if n == 1:
            self.move_disc(from_rod, to_rod)
        else:
            self.animate_hanoi(n - 1, from_rod, aux_rod, to_rod)
            self.move_disc(from_rod, to_rod)
            self.animate_hanoi(n - 1, aux_rod, to_rod, from_rod)

def main():
    root = tk.Tk()
    root.title("Tower of Hanoi")
    num_discs = 4  # You can change the number of discs here
    app = TowerOfHanoi(root, num_discs)
    root.mainloop()

if __name__ == "__main__":
    main()
```
